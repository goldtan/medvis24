```python
# charttime을 datetime 형식으로 변환
merged_df['charttime'] = pd.to_datetime(merged_df['charttime'])

# value 열의 데이터 타입 확인 및 변환
# value 열을 수치형으로 변환, 변환 불가능한 값은 NaN으로 설정
merged_df['value'] = pd.to_numeric(merged_df['value'], errors='coerce')

# NaN 값을 제거
merged_df = merged_df.dropna(subset=['value'])
```

```python
# charttime으로 정렬
merged_df = merged_df.sort_values(by='charttime')
```

```python
merged_df.to_csv("labevents.csv", index=False)
```

```python
merged_df
```

```python
subject_id = 10001884

# 특정 환자의 데이터 필터링
patient_df = merged_df[merged_df['subject_id'] == subject_id]
```

```python
# 환자의 각 검사에 대한 개별 플롯 생성
labels = patient_df['label'].unique()
labels
```

```python
# 특정 label을 입력하여 해당 검사 항목에 대한 플롯 생성

label = 'pCO2'  # 여기서 label을 입력
label_df = patient_df[patient_df['label'] == label]

plt.figure(figsize=(12, 6))
plt.plot(label_df['charttime'], label_df['value'], marker='o', label=label)
plt.title(f'Subject ID {subject_id} - {label} Results Over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
